```python
import simpy 
import random
```

```python
pista = 1

simulationTime = 60
queuePista = 0
totalTime = 0
avionsAfterInspection = 0
totalaviones = 0
cont = 0
cont1 = 0
```

```python
def tpista(env, avion):
    ale = random.uniform(0,1)
    if ale < 0.70:
        time = 1.5
        print (avion + ' es pequeño')
    else:
        time = 2.5
        print (avion + ' es grande')
    yield env.timeout(time)
    print(avion + ' se demoro %.3f minutos.' %(time))
```

```python
def llegadaavionessalida (env, pista):
    global cont
    
    
    arrivalTime = env.now
    while True:
        
        ale1 = random.uniform(0,1)
        if ale1 < 0.30:
            time = random.uniform(2,3)
        elif ale1 >=0.30 and ale1 <0.65:
            time = random.uniform(3,4)
        elif ale1 >=0.65 and ale1 <0.90:
            time = random.uniform(4,5)
        else:
            time = random.uniform(5,6)
        yield env.timeout(time)
        cont+=1
        avion= 'Avion Despegando ' + str(cont)
        env.process(customers(env, avion, pista))
        
```

```python
def llegadaavionesllegada (env, pista):
    global cont1
    
    arrivalTime = 0
    while True:
        ale = random.uniform(0,1)
        if ale < 0.30:
            time = random.uniform(2,3)
        elif ale >=0.30 and ale <0.65:
            time = random.uniform(3,4)
        elif ale >=0.65 and ale <0.90:
            time = random.uniform(4,5)
        else:
            time = random.uniform(5,6)
        yield env.timeout(time)
        cont1+=1
        a1 = 'Avion Aterrizando ' + str(cont1)
        env.process(customers(env, a1, pista))
```

```python
# Metodo global, toma el environment y los recursos del problema
def customers(env, avion, pista):
    global simulationTime, totalTime, avionsAfterInspection, avionsAfterService, queuePista, totalaviones

#Tiempo de llegada del avions
    arrivalTime=env.now
    totalaviones += 1
    print(avion + ' llega: minuto %.2f .' %(arrivalTime)) 

    
    queuePista += 1
    with pista.request() as usePista:
        yield usePista
        queuePista = queuePista - 1
        waitForInspectionTime = env.now - arrivalTime
        if waitForInspectionTime == 0:
          print(avion + ' pista: no habia fila')
        else:
          print(avion + ' para la pista: %.2f minutes in line. <-----------------' %(waitForInspectionTime))
        yield env.process(tpista(env, avion))
    pista.release(usePista)
    #Tiempo final de la primera inspeccion
    endOfInspectionTime = env.now

    exitTime = env.now
    totalTime = exitTime - arrivalTime
    print(avion + ' termina servicio: minuto %.2f' %(env.now))
    print(avion + '----------------------------------> Tiempo total %.2f' %(totalTime))
    avionsAfterService += 1
```

```python
def main1():
  print('Start of simulation')
  env = simpy.Environment()
  a = 0.9 #choose a good random seed
  random.seed(a)

  pistas = simpy.Resource(env, pista)

  env.process(llegadaavionesllegada(env, pistas))
  env.process(llegadaavionessalida(env, pistas))
  env.run(until = simulationTime)
```

```python
main1()
```
